Le document que vous consultez n'est pas une page Web statique, mais un environnement interactif appelé <strong>notebook Colab</strong>, qui vous permet d'écrire et d'exécuter du code.

## Science des données
</div>

Colab vous permet de tirer pleinement parti des bibliothèques populaires Python pour analyser et visualiser des données. La cellule de code ci-dessous utilise <strong>numpy</strong> pour générer des données aléatoires et <strong>matplotlib</strong> pour les visualiser. Pour modifier le code, cliquez simplement sur la cellule.

```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
# Charger les données TXT
with open("corpus.txt", "r", encoding="utf-8") as f:
    txt_data = f.readlines()

# Nettoyer les données (supprimer les sauts de ligne)
txt_data = [line.strip() for line in txt_data if line.strip()]

# Afficher les données
print(txt_data)
```

```
# Unifier les données JSON et TXT
unified_data = []

# Ajouter les données JSON
# Puisque json_data contient probablement des chaînes directement, nous les ajoutons telles quelles.
for item in json_data:
    unified_data.append({"text": item})  # Changé de item["text"] à item

# Ajouter les données TXT
for line in txt_data:
    unified_data.append({"text": line})

# Afficher les données unifiées
print(unified_data)
```

```
from datasets import Dataset

# Créer un dataset Hugging Face
dataset = Dataset.from_dict({"text": [item["text"] for item in unified_data]})

# Afficher le dataset
print(dataset)
```

```
from transformers import AutoTokenizer

# Charger le tokenizer
tokenizer = AutoTokenizer.from_pretrained("huggyllama/llama-7b")

# Ajouter un padding token au tokenizer
tokenizer.pad_token = tokenizer.eos_token #  Utiliser le token EOS comme token de padding
# or tokenizer.add_special_tokens({'pad_token': '[PAD]'}) # Ajouter un token de padding spécifique


# Tokeniser le dataset
def tokenize_function(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True, max_length=512)

tokenized_dataset = dataset.map(tokenize_function, batched=True)

# Afficher le dataset tokenisé
print(tokenized_dataset[0])
```

```
# Séparer en train/validation
tokenized_dataset = tokenized_dataset.train_test_split(test_size=0.1)

# Afficher les ensembles
print("Train:", tokenized_dataset["train"])
print("Validation:", tokenized_dataset["test"])
```

```
# Unifier les données
unified_data = []

# Accéder à la liste des paragraphes dans json_data
for item in json_data.get("paragraphs", []):  # Si "paragraphs" n'existe pas, utiliser une liste vide
    unified_data.append({"text": item})

# Ajouter les données TXT
for line in txt_data:
    unified_data.append({"text": line})

# Afficher les données unifiées
print(unified_data)
```

```
from transformers import AutoModelForCausalLM, AutoTokenizer, TrainingArguments, Trainer

model_name = "bigscience/bloomz-7b1-mt"  # Or try a smaller model like "bigscience/bloomz-560m"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

# Define TrainingArguments with reduced batch size, gradient accumulation and gradient checkpointing
training_args = TrainingArguments(
    output_dir="./my_llama_model",
    per_device_train_batch_size=2,  # Reduce batch size
    per_device_eval_batch_size=2,  # Reduce batch size
    gradient_accumulation_steps=2,  # Accumulate gradients over 2 steps
    gradient_checkpointing=True,    # Enable gradient checkpointing
    # ... other training arguments ...
)

# Create the Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset["train"],
    eval_dataset=tokenized_dataset["test"],
)

# Train the model
trainer.train()

# Save the model
trainer.save_model("./my_llama_model")
```
